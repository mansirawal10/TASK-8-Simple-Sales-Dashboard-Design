```python
import pandas as pd

df = pd.read_csv("Superstore.csv", encoding="latin1")  # or encoding="ISO-8859-1"

df.head()
```

**Clean the Data with Pandas**



```python
# Remove duplicates
df.drop_duplicates(inplace=True)

# Drop rows with missing critical values
df.dropna(subset=["Order Date", "Region", "Category", "Sales", "Profit"], inplace=True)

# Convert Order Date to datetime
df["Order Date"] = pd.to_datetime(df["Order Date"], errors="coerce")
df = df[df["Order Date"].notnull()]  # Remove bad dates

# Standardize text formatting
df["Region"] = df["Region"].str.strip().str.title()
df["Category"] = df["Category"].str.strip().str.title()

# Create Month-Year column
df["Month-Year"] = df["Order Date"].dt.strftime("%b-%Y")

# Save cleaned file
df.to_csv("Superstore_Sales_Clean.csv", index=False)

print("Data cleaned! Rows:", len(df))
df.head()
```
